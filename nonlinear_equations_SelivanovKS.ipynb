```python
import numpy as np

%matplotlib notebook

import matplotlib.pyplot as plt
plt.xkcd()
```

```python
def f_cos(x):
    return np.cos(x)
def f_sqrt(x):
    return np.sqrt(x)
```

```python
xx = np.linspace(0, 2)
fig, ax = plt.subplots()
ax.axvline(0, ls='--', lw=1); plt.axhline(0, ls='--', lw=1)
ax.plot(xx, f_cos(xx))
ax.plot(xx, f_sqrt(xx))
```

```python
def fi(a, E):
    count = 0
    x = 0.5
    x1 = x - a * (np.sqrt(x) - np.cos(x))
    while abs(x1 - x) > E:
        count += 1
        x = x1
        x1 = x - a * (np.sqrt(x) - np.cos(x))
    return count, x
```

```python
E = 0.001
a = 1
count, x = fi(a, E)
print('Number of iterations =', count)
print('Solution =', x)
```
