```python
import numpy as np

%matplotlib notebook

import matplotlib.pyplot as plt
plt.xkcd()
```

```python
def f_cos(x):
    return np.cos(x)
def f_sqrt(x):
    return np.sqrt(x)
```

```python
xx = np.linspace(0, 2)
fig, ax = plt.subplots()
ax.axvline(0, ls='--', lw=1); plt.axhline(0, ls='--', lw=1)
ax.plot(xx, f_cos(xx))
ax.plot(xx, f_sqrt(xx))
```

```python
def fi(a, E):
    count = 0
    x = 0.5
    x1 = x - a * (np.sqrt(x) - np.cos(x))
    while abs(x1 - x) > E:
        count += 1
        x = x1
        x1 = x - a * (np.sqrt(x) - np.cos(x))
    return count, x
```

```python
E = 0.001
a = 1
count, x = fi(a, E)
print('Number of iterations =', count)
print('Solution =', x)
```

```python
def f_complex(z):
    return z**3 - 1
def f_complex_prime(z):
    return 3 * z**2
```

```python
def newton_raphson(z, E):
    while abs(f_complex(z)) > E:
        z = z - f_complex(z)/f_complex_prime(z)
    return z
```

```python
E = 0.001
z1 = 1+1j
z2 = -1+1j
z3 = -1-1j
z1 = newton_raphson(z1, E)
z2 = newton_raphson(z2, E)
z3 = newton_raphson(z3, E)

print('z1 =', z1)
print('z2 =', z2)
print('z3 =', z3)
```

```python
fig, ax = plt.subplots()
ax.axvline(0, ls='--', lw=1); plt.axhline(0, ls='--', lw=1)
ax.plot()
ax.plot(z1.real, z1.imag, color='red', marker='o')
ax.plot(z2.real, z2.imag, color='green', marker='o')
ax.plot(z3.real, z3.imag, color='blue', marker='o')
```
